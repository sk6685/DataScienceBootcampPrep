```python
import pandas as pd
```

```python
df = pd.read_csv('Consumer_Complaints.csv')
```

```python
df = df[['Date received','Company','Product','State','Issue',]]
```

```python
df=df.rename(columns = {'Date received':'Year'})
```

```python
print(df['Company'].value_counts())
```

```python
print(df[1:100])
```

```python
df['Year'] = pd.to_datetime(df['Year']).dt.year
print(df.Year)
```

```python
df.sort_values(by=['Year'])
```

```python
import numpy as np
from matplotlib import pyplot as plt

plt.style.use('ggplot')
y_axis = df["Year"].value_counts()
plt.bar(range(len(y_axis)), y_axis)
xtickloc = np.array(range(len(y_axis)))
plt.xticks(xtickloc, y_axis.index, rotation=90)
plt.title("")
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')

y_axis = df["Product"].value_counts()
plt.bar(range(len(y_axis)), y_axis)
xtickloc = np.array(range(len(y_axis)))
plt.xticks(xtickloc, y_axis.index, rotation=90)
plt.title('Summary of Products involving complaints')
plt.show()
```

```python
Comp_data=df.groupby('Company').size().nlargest(15)
print(Comp_data)
Comp_data.values
```

```python
import matplotlib.pyplot as plt;
import numpy as np
import matplotlib.pyplot as plt
 
objects = Comp_data.index
y_pos = np.arange(len(Comp_data))
values = Comp_data.values

 
plt.barh(y_pos, values, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Number of Complaints')
plt.title('TOP 15 Companies with more number of Complaints')
 
plt.show()
```
